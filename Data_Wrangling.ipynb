## Manos a la Obra


Importar el modulo pandas.


```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import numpy as np
```

Cargar el conjunto de datos en el dataframe


```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv")
```

## Encontrar Duplicados


En esta sección identificará valores duplicados en el conjunto de datos


Encuentre cuantas filas duplicadas existen en el dataframe.


```python
# su código va aquí
df.duplicated().sum()
```

## Borrar Duplicados


Retirar las filas duplicadas del dataframe.


```python
# su código va aquí
no_duplicates= df.drop_duplicates()
no_duplicates
```

Verifique si los valores duplicados ya no están.


```python
# su código va aquí
no_duplicates.duplicated().sum()
```

## Encontrar Valores Faltantes


Encuentre los valores faltantes en todas las columnas.


```python
# su código va aquí
missing_data = no_duplicates.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Encuentre cuantas filas faltan en la columna 'WorkLoc'


```python
# su código va aquí
no_duplicates['WorkLoc'].isnull().sum()
```

## Ingresar los Valores Faltantes


Encuentre el valor de conteo para la columna WorkLoc.


```python
# su código va aquí
no_duplicates['WorkLoc'].count()
```

Identifique el valor mas frecuente (mayoritario) en la columna WorkLoc


```python
# haga una anotación del valor mayoritario aquí, para futuras referencias
no_duplicates['WorkLoc'].value_counts()
```

Ingrese (reemplace) todas las filas vacías en la columna WorkLoc con el valor que ha identificado como mayoritario.


```python
# su código va aquí
no_duplicates['WorkLoc'].replace(np.nan,'Office', inplace=True)
```

Después, idealmente no debería haber filas vacías en la columna WorkLoc


Verifique si el proceso fue exitoso.


```python
# su código va aquí
print(no_duplicates['WorkLoc'].isnull().sum())
print('\n')
print(no_duplicates['WorkLoc'].value_counts())
```

## Normalizar los Datos


Hay dos columnas en el conjunto de datos que hablan sobre compensación.

Una es "CompTotal". Esta columna habla sobre que tan a menudo se le paga al desarrollador (Anualmente, Mensualmente, Semanalmente).

La otra es  "CompTotal". Esta columna habla sobre cuanto se le paga al desarrollador por Año, Mes o Semana dependiendo de su "CompFreq". 

Esto hace difícil comparar el total de compensación de los desarrolladores.

En esta sección usted creará  una nueva columna llamada 'NormalizedAnnualCompensation' que contendrá la 'Annual Compensation' ( Compensación Anual) sin considerar 'CompFreq'.

Una vez que la columna este terminada, hará la comparación de salarios sencilla.


<hr>


Listar las diversas categorias en la columna 'CompFreq' 


```python
# su código va aquí
categ = no_duplicates['CompFreq'].unique()
categ
```

Crear una columna nueva llamada 'NormalizedAnnualCompensation'. Utilice la pista dada abajo si es necesario.


Haga Doble clic para ver la **Pista**.



Utilice la siguiente lógica para llegar a los valores de la columna NormalizedAnnualCompensation.

Si CompFreq es Yearly entonces utilice el valor existente en CompTotal
Si CompFreq es Monthly entonces multiplique el valor en CompTotal por 12 (meses en el año)
Si CompFreq es Weekly entonces multiplique el valor en CompTotal por 52 (semanas en el año)


```python
no_duplicates = no_duplicates.dropna()
```

```python
Weekly = no_duplicates[(no_duplicates.CompFreq == "Weekly") & (no_duplicates.CompTotal)] 
w = 52*Weekly['CompTotal']

Monthly = no_duplicates[(no_duplicates.CompFreq == "Monthly") & (no_duplicates.CompTotal)] 
m = 12*Monthly['CompTotal']

Annual= no_duplicates[(no_duplicates.CompFreq == "Yearly") & (no_duplicates.CompTotal)] 
y = 1*Annual['CompTotal']

NormalizedAnnualCompensation = pd.concat([w,m,y])

no_duplicates['NormalizedAnnualCompensation'] = NormalizedAnnualCompensation

no_duplicates
```
